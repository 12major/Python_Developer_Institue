```python
import random

class Airplane():
    
    def __init__(self,
                model,
                manufacturer,
                seating_capacity,
                max_speed,
                fuel_capacity,
                crash_probability=10,
                fuel_consomption=.3):
        
        self.model              = model
        self.manufacturer       = manufacturer
        self.seating_capacity   = seating_capacity
        self.max_speed          = max_speed
        self.fuel_capacity      = fuel_capacity
        self.crash_probability  = crash_probability
        self.fuel_consommation  = fuel_consomption
        
        self.current_fuel       = 0
        self.current_passengers = {}
        self.current_seat       = 0
    
    def add_passenger(self, passenger):
        if len(self.current_passengers)+1 > self.seating_capacity:
            print("Sorry, this plane is full.")
        else:
            self.current_passengers[self.current_seat] = passenger
            print("Passenger {} has been added to place {}".format(passenger.name, self.current_seat))
            self.current_seat += 1
        
    def fill_plane(self, quantity_to_fill, price_per_liter):
        
        available_place = self.fuel_capacity - self.current_fuel
        if quantity_to_fill > available_place:
            quantity_to_fill = available_place
        
        gas_price = quantity_to_fill*price_per_liter
        self.current_fuel += quantity_to_fill
        
        print("The airplane has been filled, he got now {} liters of fuel, it cost you {} dollars".format(self.current_fuel,
                                                                                                         gas_price))
        
        return gas_price
    
    def take_off(self):
        if random.randint(1,self.crash_probability) == 1:
            print("Your plane just crashed ;(")
            return False
        else:
            print("Took off ! Good chance for the landing")
            return True
        
    def land(self):
        if random.randint(1,self.crash_probability) == 1:
            print("Your plane just crashed ;(")
            return False
        else:
            print("ClapClapClap")
            return True
    
    def travel(self, departure, destination, distance):
        
        fuel_needed = distance*self.fuel_consumption
        if fuel_needed > self.current_fuel:
            print("You don't have enough fuel..")
            return False
        
        if not self.take_off() or not self.land():
            return False
        
        print("Mazal tov, you are arrived to {}".format(destination))
        self.current_fuel -= fuel_needed
        
class Passenger:
    
    def __init__(self,
                name,
                weight):
        
        self.name   = name
        self.weight = weight        
```

#### Step 4:

Create a `play` function, it should run a Roulette game between the players, here are the rules:
-  On the wheel, there are numbers from 0 to 36
-  Each player choose a bet: it can be a single number or it can be ODD or EVEN
-  We roll a random number between 0 and 36, if the player wins a bet on a single number, then he wins 35 times his bet, if he wins a bet on ODD or on EVEN, he wins 2 times his bet<br>
Player can quit at any moment of the game if they want or if they don't have money.<br>

#### Step 5:

```python
import random

class RouletteGame:
    
    def __init__(self,
                entry_cost,
                seating_capacity):
        
        self.entry_cost = entry_cost
        self.seating_capacity = seating_capacity
        
        self.total_gain = 0
        self.current_players = []
    
    def take_money_from_player(self, player, amount):
        player.total_money -= amount
        self.total_gain    += amount
    
    def give_money_to_player(self, player, amount):
        self.total_gain    -= amount
        player.total_money += amount
        
    def add_player(self,
                    player):
        if player.total_money >= self.entry_cost and len(self.current_players) < self.seating_capacity:
            self.current_players.append(player)
            self.take_money_from_player(player, self.entry_cost)
        else:
            print("Sorry, you can't enter that table.")
    
    def play(self):
        players_bets = {}
        for player in self.current_players:
            player_bet = int(input("Hey, {}, choose a number\n>>> ".format(player.name)))
            players_bets[player] = player_bet
            #player is <__main__.Object at 0xf54ed3>
        wheel_roll = random.randint(0,36)
        for player in self.current_players:
            if players_bets[player] == wheel_roll:
                print("Congrats, {}, you just won 35$".format(player.name))
                self.give_money_to_player(player, 35)
            else:
                print("Sorry {}, you lose your dollar".format(player.name))
                self.take_money_from_player(player, 1)
        
class Player:
    
    def __init__(self,
                name,
                total_money):
        self.name = name.capitalize()
        self.total_money = total_money
        
        
my_game = RouletteGame(entry_cost=10, seating_capacity=3)
player1 = Player("eyal", 100)
player2 = Player("John",150)
my_game.add_player(player1)
my_game.add_player(player2)
while True:
    my_game.play()
    again = input("Play again??[y/n]\n>>> ")
    if again.lower() != "y":
        break
```
